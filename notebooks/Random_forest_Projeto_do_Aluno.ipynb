## Random Forest

```python
import pandas as pd
import numpy as np
from sklearn import tree as tr
import cv2
import pydotplus
import matplotlib.pyplot as plt
from IPython.display import Image
from six import StringIO
from sklearn import metrics as mt
from sklearn.linear_model import LogisticRegression as lr
from sklearn.ensemble import RandomForestClassifier
```

```python
X_training = pd.read_csv('/content/X_training.csv')
y_training = pd.read_csv('/content/y_training.csv')
X_test = pd.read_csv('/content/X_test.csv')
y_test = pd.read_csv('/content/y_test.csv')
X_validation = pd.read_csv('/content/X_validation.csv')
y_validation = pd.read_csv('/content/y_validation.csv')
```

##Metricas para os dados do Treino

```python
model = tr.DecisionTreeRegressor()
model.fit(X_training,y_training)
yhatTrain = model.predict(X_training)
```

```python
acctrain = mt.accuracy_score(y_training, yhatTrain)
print('A Acuracia é: {:}'.format(acctrain))
```

```python
precision_train = mt.precision_score(y_training, yhatTrain)
print('A Precisião é: {:}'.format(precision_train))
```

```python
recall_train = mt.recall_score(y_training, yhatTrain)
print('O recall é: {:}'.format(recall_train))
```

```python
f1_train = mt.recall_score(y_training, yhatTrain)
print('O recall é: {:}'.format(f1_train))
```

```python
Classif_train = mt.classification_report(y_training, yhatTrain)
print(Classif_train)
```
